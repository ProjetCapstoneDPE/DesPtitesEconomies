```python
region_code='PACA'
fname = f'dpe_{region_code}.csv'

region = ['04','05','06','13','83','84']

df_region = pd.DataFrame(data=None, index=None, columns=None, dtype=None, copy=None)

for i in range(len(region)):
    file = f'DPE_dep_{region[i]}.csv'
    df=pd.read_csv(file)
    print(len(df))
    df_region= pd.concat([df_region,df], ignore_index=True)
```

```python
df_region.to_csv(fname,index=False)
```
